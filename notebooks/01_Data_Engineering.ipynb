```python
# -------------------------------------------------------------
# Step 2: Load the raw dataset
# -------------------------------------------------------------
print("📥 Loading dataset...")

df = pd.read_csv("../data/raw/AB_NYC_2019.csv")

print("✅ Dataset loaded successfully!")
print("Initial data shape:", df.shape)
print("\n🔍 Preview of data:")
print(df.head())
```

```python
# -------------------------------------------------------------
# Step 3: Understand structure — schema, datatypes, nulls
# -------------------------------------------------------------
print("\n📊 Data types and null value check:")
print(df.info())
```

```python
# -------------------------------------------------------------
# Step 4: Check for missing values
# -------------------------------------------------------------
print("\n❓ Any missing values?")
print(df.isnull().sum())
```

```python
#-------------------------------------------------------------
# Step 7: Drop unnecessary columns
#-------------------------------------------------------------
#Drop columns named "name", "host_name", and "last_review"
df = df.drop(columns=['id','name','host_id', 'host_name', 'last_review'])

#drop NaN values in the dataset
df = df.dropna()
df.head()
```

```python
# -------------------------------------------------------------
# Step 8: Basic stats check (spot outliers or issues)
# -------------------------------------------------------------
print("\n📈 Basic summary stats for numeric columns:")
print(df.describe())
```

```python
# -------------------------------------------------------------
# Step 10: Drop duplicates if any
# -------------------------------------------------------------
print("\n🧹 Checking for duplicate rows...")

duplicates = df.duplicated().sum()
if duplicates > 0:
    print(f"⚠️ Found {duplicates} duplicate rows. Dropping them...")
    df = df.drop_duplicates()
else:
    print("✅ No duplicate rows found.")
```

```python
# -------------------------------------------------------------
# Step 13: Save the cleaned version to processed folder
# -------------------------------------------------------------
print("\n💾 Saving cleaned version to interim file (optional step)...")
df.to_csv("../data/processed/AB_NYC_2019_working_table.csv", index=False)
print("✅ Cleaned data saved to '../data/processed/AB_NYC_2019_working_table.csv'")
```
